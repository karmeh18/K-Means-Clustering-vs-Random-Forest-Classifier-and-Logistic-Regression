```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler=StandardScaler()
le=LabelEncoder()
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import warnings
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
warnings.filterwarnings("ignore")
```

# Breast Cancer Dataset

```python
data=pd.read_csv("C:\\Users\\Karan\\Desktop\\Data\\breast_cancer_data\\breast_disease_data.csv").iloc[:,1:]
data
```

```python
data["diagnosis"]=le.fit_transform(data["diagnosis"])
data.head()
```

# Train-Test-Split

```python
X=data.drop(columns=["diagnosis"])
X
y=data["diagnosis"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# LogisticRegression

```python
log_model=log.fit(X_train,y_train)
log_predict=log_model.predict(X_test)
print("Accuracy on Test Data with Logistic Regression is: ",accuracy_score(y_test,log_predict))

log_model1=log.fit(X_test,y_test)
log_predict1=log_model1.predict(X_train)
print("Accuracy on Training data with Logistic Regression is: ",accuracy_score(y_train,log_predict1))
```

# Random Forest Classifier

```python
rfc_model=rfc.fit(X_train,y_train)
rfc_predict=rfc_model.predict(X_test)
print("Accuracy on Test Data with Logistic Regression is: ",accuracy_score(y_test,rfc_predict))

rfc_model1=rfc.fit(X_test,y_test)
rfc_predict1=rfc_model1.predict(X_train)
print("Accuracy on Training data with Logistic Regression is: ",accuracy_score(y_train,rfc_predict1))
```

# K-Means Clustering

```python
X_train_new=X_train.copy()
X_train_new
X_test_new=X_test.copy()
X_test_new
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,random_state=42)
kmeans_predict_train=kmeans.fit_predict(X_train_new)
kmeans_predict_test=kmeans.fit_predict(X_test_new)
```

```python
print("Accuracy on Training Data with K-Means is: ",accuracy_score(y_train,kmeans_predict_train))
print("Accuracy on Training Data with K-Means is: ",accuracy_score(y_test,kmeans_predict_test))
```

# K-Means Clustering with Standardized Data

```python
X_train_new_stan=pd.DataFrame(scaler.fit_transform(X_train_new),columns=X_train.columns)
X_test_new_stan=pd.DataFrame(scaler.fit_transform(X_test_new),columns=X_train.columns)
X_test_new_stan
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,random_state=42)
kmeans_predict_train=kmeans.fit_predict(X_train_new_stan)
kmeans_predict_test=kmeans.fit_predict(X_test_new_stan)
```

```python
print("Accuracy on Training Data with K-Means is: ",accuracy_score(y_train,kmeans_predict_train))
print("Accuracy on Training Data with K-Means is: ",accuracy_score(y_test,kmeans_predict_test))
```

## It is possible to use K-Means to generate predictions in certain situations. For example, after clustering the data you can use the cluster assignments as an additional feature in a supervised learning model such as Random Forest Classifier. Or in a semi-supervised learning scenario, where you have labelled data points & unlabelled data points & propagate the labels from the labelled data points to the closest cluster center, effectively generating predictions for the unlabelled data points

## It's important to note that K-Means is not a classification algorithm, it's a clustering algorithm & it's primary goal is not to predict a target variable. It's better to use a supervised algorithm like Random Forest Classifier for classification task.
